(a) Replace the values of a column

<font size = "3">

- Import the "data_raw/sprint_results.csv" file
- Print the list with the unique values that are not numeric in the column "positionText"
- Replace the non-numerica values ("R" and "N") with ```np.nan``` (missing value) in the column "positionText"
- Create a new column "positionNumeric" that converts the modified "positionText" column from string to numeric using ```pd.to_numeric()```

HINT: See the example in Lecture 12

````python
# Write your own code
sprint_results = pd.read_csv("data_raw/sprint_results.csv")
#display(sprint_results) check before
positionTextNoNum = sprint_results.query("positionText.str.isnumeric() == False")
#display(positionTextNoNum) check after
uniPositionText = pd.unique(positionTextNoNum["positionText"])
print(uniPositionText)

# replace R and N
oldList = uniPositionText
sprint_results["positionText"] = sprint_results["positionText"].replace(oldList, np.nan)

sprint_results["positionNumeric"] = pd.to_numeric(sprint_results["positionText"])
display(sprint_results) # check afterwards
````

(b) Recode a numeric colum

<font size = "3">

- Import the "data_raw/races.csv" file
- Recode the "year" column into a new column "year_brackets" with the following categories

$\qquad$ ``` ["1950-1974","1975-1999","2000-onwards"] ```

- Display the new "year_brackets" variable

$\qquad$ HINT: Use the "pd.cut()" command. See Lecture 12 (check the note in the example!)

````python
# Write your own code
raceRaw = pd.read_csv("data_raw/races.csv")

labelsYear = ["1950-1974","1975-1999","2000-onwards"]
binsYear = [1949,1974,1999,1000000]
raceRaw["year_brackets"] = pd.cut(raceRaw["year"], bins = binsYear, right = True, labels = labelsYear)

display(raceRaw["year_brackets"])
````

(c) Aggregate and query

<font size = "3">

- Import the "data_raw/sprint_results.csv" file
- Subset the observations for 'constructorId == 3' using ```.query()```
- Obtain the sum of "laps" by "driverId" using ``` .groupby().agg() ```
- Subset the drivers with a sum of laps $\ge 40$
- Store the results of all these operations in a dataframe ```sprint_results_agg ```
- Display the ```sprint_results_agg``` dataframe
- Provide a one-sentence explanation of the information in the ```sprint_results_agg ``` table

````python
# Write your own code
cSprintResults = pd.read_csv("data_raw/sprint_results.csv")


subsetConstructorID = cSprintResults.query("constructorId == 3")
display(subsetConstructorID.sort_values("raceId",ascending = True))
aggDriverId = subsetConstructorID.groupby("driverId").agg(sumLaps = ("laps","sum"))
display(aggDriverId)

sprint_results_agg = aggDriverId.query("sumLaps >= 40")
display(sprint_results_agg)

print("Explanation: sprint_results_agg shows the driverId and sum of their laps for those with constructorId of 3 and at least 40 total laps. ")
````

(d) Aggregate and sort

<font size = "3">

- Import the "data_raw/sprint_results.csv" file
- Obtain the sum of "laps" by "driverId" using ``` .groupby().agg() ```
- Sort the teams in descending order using of the sum of laps using ``` .sort_values() ```
- Display your results

````python
# Write your own code
dSprintResults = pd.read_csv("data_raw/sprint_results.csv")
sumLapsDriverId = dSprintResults.groupby("driverId").agg(sumLaps = ("laps","sum"))
sumLapsDriverId = sumLapsDriverId.sort_values("sumLaps", ascending = False)
display(sumLapsDriverId)
````
